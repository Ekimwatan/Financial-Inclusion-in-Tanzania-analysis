```python
import geopandas as gpd
import pandas as pd
import folium
from folium import Choropleth, Circle, Marker, GeoJson
from folium.plugins import HeatMap, MarkerCluster
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers= pd.read_csv('training.csv')
```

```python
print("There are {} customer records".format(len(customers)))
```

```python
mobile_money=pd.read_csv('FSDT_FinAccessMapping/mobilemoney_agents_for_upload_win.csv', encoding='latin1')
```

```python
print("There are {} mobile money agents total".format(len(mobile_money)))
```

```python
column_names=['Unique_respondent_ID', 'Age', 'gender', 'Marital_status', 'Highest_level_of_education'
             ,'own_land', 'own_certified_land', 'own_mobile_phone', 'salaries/wages', 'trading', 
             'providing_service', 'casual_labor', 'rental_income', 'earning_interest', 'pension',
             'social_welfare', 'get_upkeep_from_someone', 'expenses_paid_for_by_someone', 'other', 
             'employer', 'trading_goods_type', 'service_type', 'sent_money_in_past_12months', 
             'last_sent_money', 'received_money_in_last_12_months', 'last_money_received', 
              'mobile_money_usage', 'bills_payment_by_mobile_money', 'kiswahili_literacy', 
              'English_literacy', 'Latitude', 'Longitude', 'mobile_money', 'save', 'borrow', 
              'insurance', 'mobile_money_classification']
```

```python
customers.columns=column_names
```

```python
print(customers.columns)
```

```python
customers.head(10)
```

```python
#Create a GeoDataFrame
customers_gdf= gpd.GeoDataFrame(customers, geometry=gpd.points_from_xy(customers.Longitude, customers.Latitude))
customers_gdf.crs = {'init': 'epsg:32737'}
```

```python
customers_gdf.head()
```

```python
mobile_money_gdf=gpd.GeoDataFrame(mobile_money, geometry=gpd.points_from_xy(mobile_money.longitude, mobile_money.latitude))
## Set the coordinate reference system (CRS) to EPSG 4326
mobile_money_gdf.crs = {'init': 'epsg:32737'}
```

```python
mobile_money_gdf.head()
```

```python
#Get Wards in Tanzania data
tz_wards=gpd.read_file('tza_admbnda_adm3_20181019/tza_admbnda_adm3_20181019.shp')

tz_wards.head()
```

```python
#Spatial Join Customers and Ward data
customers_inwards= gpd.sjoin(customers_gdf, tz_wards)
```

```python
customers_inwards.head()
customers_inwards['ADM2_EN'].value_counts()
```

```python
customers_inwards['ADM2_EN'].value_counts()
```

```python
#Select Wawi Ward to work with. 

district='Kinondoni'
```

```python
customers_in_Tunduma=customers_inwards[customers_inwards['ADM2_EN']==district]
customers_in_Tunduma.columns
```

```python
len(customers_in_Tunduma)
```

```python
mobile_money_in_Tunduma=mobile_money_gdf[mobile_money_gdf['district']==district]
```

```python
len(mobile_money_in_Tunduma)
```

```python
#Putting a 2km buffer on mobile money locations
two_mile_buffer = mobile_money_in_Tunduma.geometry.buffer(1.5*5280)
two_mile_buffer.head()
```

```python
m10 = folium.Map(location=[-5.544751, 35.435301], zoom_start=11)
HeatMap(data=customers_in_Tunduma[['Latitude', 'Longitude']], radius=15).add_to(m10)
for idx, row in mobile_money_in_Tunduma.iterrows():
    Marker([row['latitude'], row['longitude']]).add_to(m10)
    
#folium.GeoJson(two_mile_buffer).add_to(m10)
m10
```

```python
#HeatMap of customers in Tanzania
m20 = folium.Map(location=[-5.544751, 35.435301], zoom_start=11)
HeatMap(data=customers[['Latitude', 'Longitude']], radius=9).add_to(m20)
m20
```

```python
m30 = folium.Map(location=[-5.544751, 35.435301], zoom_start=11)
HeatMap(data=mobile_money[['latitude', 'longitude']], radius=9).add_to(m30)
m30
```

```python
coverage = gpd.GeoDataFrame(geometry=mobile_money_gdf.geometry).buffer(1000)
my_union = coverage.geometry.unary_union
outside_range = customers_gdf.loc[~customers_gdf["geometry"].apply(lambda x: my_union.contains(x))]
```

```python
outside_range.head()
```

```python
m40= folium.Map(location=[-5.544751, 35.435301], zoom_start=11)
for idx, row in outside_range.iterrows():
    Marker([row['Latitude'], row['Longitude']]).add_to(m40)
    
m40

```
